```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Year=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
Temp_I=[0.1,8.3,2.3,4,5,22,12.4,9.8,3.4,2.4,2]
```

```python
plt.plot(Year,Temp_I)
plt.xlabel('year')
plt.ylabel('tempindex')
plt.title('globalwarming',{'fontsize':12,'horizontalalignment':'center'})
plt.show()
```

```python
Month=['jan','feb','march','april','may','june','july','august','september','october','november','december']
customer1=[10,12,22,11,22,2,33,23,21,12,44,17]
customer2=[1,22,32,41,21,25,13,3,11,15,14,5]
```

```python
plt.plot(Month,customer1,color='red',label='cust1',marker='o')
plt.plot(Month,customer2,color='blue',label='customer2',marker='^')
plt.xlabel('month')
plt.ylabel('elect consumption')
plt.title("buildind details")
plt.legend(loc='lower left')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(Month,customer1,color='red',label='cust1',marker='o')
plt.xlabel('month')
plt.ylabel('elect consumption')
plt.title("buildind details1")
plt.subplot(1,2,2)
plt.plot(Month,customer2,color='blue',label='customer2',marker='^')
plt.xlabel('month')
plt.ylabel('elect consumption')
plt.title("buildind details2")
plt.show()
```

```python
plt.scatter(Month,customer1,color='red',label='cust1')
plt.scatter(Month,customer2,color='blue',label='cust2')
plt.xlabel('month')
plt.ylabel('elec consumption')
plt.title("scatter plot of building consumption")
plt.grid()
plt.legend(loc="best")
plt.show()
```

```python
plt.hist(customer1,bins=100,color='green')
plt.xlabel('month')
plt.ylabel('elec consumption')
plt.title(" histogram")
plt.show()
```

```python
plt.bar(Month,customer1,color='blue',width=0.8)
plt.show()
```

```python
plt.bar(Month,customer1,color='red',label='cust1')
plt.bar(Month,customer2,color='blue',label='cust2')
plt.xlabel('month')
plt.ylabel('elec consumption')
plt.title("bar chart")
plt.legend(loc="best")
plt.show()
```

```python
bar_width=0.4
month_b=np.arange(12)
plt.bar(month_b,customer1,bar_width,color='blue',label="customer1")
plt.bar(month_b+bar_width,customer2,bar_width,color='red',label="customer2")
plt.xlabel('month')
plt.ylabel('elec consumption')
plt.title("bar chart")
plt.legend(loc="best")
plt.show()
```

```python
bar_width=0.4
month_b=np.arange(12)
plt.bar(month_b,customer1,bar_width,color='blue',label="customer1")
plt.bar(month_b+bar_width,customer2,bar_width,color='red',label="customer2")
plt.xlabel('month')
plt.ylabel('elec consumption')
plt.title("bar chart")
plt.legend(loc="best")
plt.xticks(month_b+(bar_width)/12,('jan','feb','march','april','may','june','july','august','september','october','november','december'))
plt.show()
```
