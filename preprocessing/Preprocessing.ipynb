```python
"Concatination"
newcol=pd.read_csv('https://raw.githubusercontent.com/abrahul/Data_Science_step_by_step/main/preprocessing/DataNew.csv')
newcol.head()
```

```python
dataset8=dataset8.apply(pd.to_numeric)
dataset10=pd.concat([dataset8,newcol],axis=1)
dataset10.info()
```

```python
dataset10.head()
```

```python
dataset11=pd.get_dummies(dataset10)
dataset11.info()
```

```python
dataset11.head()
```

```python
"Normalization"
from sklearn.preprocessing import minmax_scale,normalize
#firt using min-max method
dataset12=minmax_scale(dataset11,feature_range=(0,1))
pd.DataFrame(dataset12)
```

```python
#Normalised method
dataset13=normalize(dataset11,norm='l2',axis=0)
#axis=0 to normalize feature/column,axis=1 to normalise sample(row)
#l1=manhattan,l2=euclidrean
pd.DataFrame(dataset13)
```

```python
dataset14=pd.DataFrame(dataset13,columns=['Time','E_plug','Ehit','price','temp','offpeak','peak'])
dataset14
```
